### Exercise 6: Image Transformation

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("images/flower.jpg")

# Change color to RGB (from BGR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
### Averaging Blur
```

```python
avr_blurred = cv2.blur(img_rgb, (11,11),0)
plt.imshow(avr_blurred)
```

## Gaussian Blur

```python
#
Gauss_blurred = cv2.GaussianBlur(img_rgb, (11,11),0)
plt.imshow(Gauss_blurred)
```

### Median Blur 

```python
median_blurred = cv2.medianBlur(img_rgb, 11)
plt.imshow(median_blurred)
```

```python
### Show all the smoothings

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Average Blur')
ax1.imshow(avr_blurred)

ax2.set_title('Gaussian Blur')
ax2.imshow(Gauss_blurred)

ax3.set_title('Median Blur')
ax3.imshow(median_blurred)
```

#### Exercise 7 c: Image Sharpening

```python
img_eye_bgr = cv2.imread("images/eye.jpg")

# Change color to RGB (from BGR)
img_eye = cv2.cvtColor(img_eye_bgr, cv2.COLOR_BGR2RGB)
#plt.imshow(img_eye)
```

```python
# Using sharpening filter

sharpening_kernel = np.array([[0, -1, 0],
                             [-1, 5, -1],
                             [0, -1, 0]])
sharpened_image = cv2.filter2D(img_eye, -1, sharpening_kernel)

#plt.imshow(sharpened_image)
```

```python
### Show all the images

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original')
ax1.imshow(img_eye)

ax2.set_title('Sharpened')
ax2.imshow(sharpened_image)
```

```python
# option 2: Using Unsharp masking
gauss_blur = cv2.GaussianBlur(img_eye,(5,5),10)
sharpened = cv2.addWeighted(img_eye,1.5,gauss_blur,-0.5,0)
```

```python
### Show all the images

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original')
ax1.imshow(img_eye)

ax2.set_title('Sharpened')
ax2.imshow(sharpened)
```
