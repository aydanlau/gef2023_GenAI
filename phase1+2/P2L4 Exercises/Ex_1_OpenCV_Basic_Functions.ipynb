### Exercise 1: OpenCV Basic Functions

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### 1a. Reading the Image

```python
img = cv2.imread("images/flower.jpg")
```

### 2b. Displaying the Image

```python
plt.imshow(img)
```

### Note: OpenCV uses by default the BGR pixel format. The most common standard for computers to represent colors is RGB. It's simple to convert from BGR to RGB.

### 2. Conversion to RGB from BGR

```python
# Convert BGR to RGB color space
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(imgRGB)
```

### 3. Conversion of RGB to Grayscale

```python
# convert image to grayscale
gray = cv2.cvtColor(imgRGB, cv2.COLOR_RGB2GRAY)
# show image
plt.imshow(gray, cmap='gray')
```

### 4. Saving the Image 

```python
saveImg = cv2.imwrite("gray.jpg", gray)
```

```python
# saving an image in a different format
imgPng = cv2.imwrite("gray.png", gray)
```

### 5a. Extract and merge the color components

```python
# Isolate RGB channels

r,g,b = cv2.split(imgRGB)  #or
#r,g,b = imgRGB [ height, width]

# Visualize the individual color channels
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='gray')
ax2.set_title('G channel')
ax2.imshow(g, cmap='gray')
ax3.set_title('B channel')
ax3.imshow(b, cmap='gray')
```

### 5b.  Merge the color components

```python
image2 = cv2.merge((r,g,b))
plt.imshow(image2)
```
