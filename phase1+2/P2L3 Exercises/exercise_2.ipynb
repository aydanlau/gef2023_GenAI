```python
# padding np.pad(array, (top, bottom), (left, right), mode, values)

white = [255,255]
sanka_img_pad = np.pad(sanka_img0, ((100,100), (100,100),(0,0)), mode='constant', constant_values=white)

plt.imshow(sanka_img_pad)
```

```python
# e) Show the color components
red = sanka_img0[:,:,0]
green = sanka_img0[:,:,1] 
blue = sanka_img0[:,:,2]

plt.imshow(red, cmap='gray')

# Alternatively

#####
#red, green, blue = sanka_img0.copy(), sanka_img0.copy(), sanka_img0.copy()
#red[:, :, (1, 2)] = 0
#green[:, :, (0, 2)] = 0
#blue[:, :, (0, 1)] = 0

#img_rgb = np.concatenate((red,green,blue), axis=1)
#plt.figure(figsize=(15, 15))
#plt.imshow(img_rgb)
####
```

```python
#recover the original image
# Using NumPy dstack
rgb_image = np.dstack((red, green, blue))
plt.imshow(rgb_image)
```

```python
#f) Color reduction

img_0 = (sanka_img0 // 64) * 64
img_1 = (sanka_img0 // 128) * 128
img_all = np.concatenate((sanka_img0, img_0, img_1), axis=1)
plt.figure(figsize=(15, 15))
plt.imshow(img_all)
```

```python
#i
src = np.array(Image.open('images/Sanka_img.jpg').resize((128, 128)))
dst = np.array(Image.open('images/Sanka_img.jpg').resize((256, 256))) // 8
dst_copy = dst.copy()
dst_copy[64:128, 128:192] = src[32:96, 32:96]
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(src)
plt.title('Original')
fig.add_subplot(1, 2, 2)
plt.imshow(dst_copy)
plt.title('Pasted with slice')
```

```python
# Binarization
img = np.array(Image.open('images/Sanka_img.jpg'))
img_64 = (img > 64) * 255
img_128 = (img > 128) * 255
fig = plt.figure(figsize=(15, 15))
img_all = np.concatenate((img, img_64, img_128), axis=1)
plt.imshow(img_all)
```

```python
# Flip
img = np.array(Image.open('images/Sanka_img.jpg'))
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(np.flipud(img))
fig.add_subplot(1, 2, 2)
plt.imshow(np.fliplr(img))
```

```python
# Color blending

img = np.array(Image.open('images/cat.jpg'))
img0 = np.array(Image.open('images/mountain.jpg').resize(img.shape[1::-1])) # resize takes 2 arguments (WIDTH, HEIGHT)
img.shape
img_blend = (img * 0.7 + img0 * 0.3).astype(np.uint8)   # Blending them in
plt.figure(figsize=(10, 10))
plt.imshow(img_blend)
```
