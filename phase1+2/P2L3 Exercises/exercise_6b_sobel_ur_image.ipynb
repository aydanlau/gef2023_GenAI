```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# generate the sobel filters  
my_img = plt.imread("images/my_img.png")
my_img_gray = np.dot(my_img[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(my_img_gray, cmap='gray')
```

```python
# x-kernel 
sobelx = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])

#y-kernel
sobely = np.array([[-1, -2, -1],[ 0,  0,  0],[ 1,  2,  1]])
```

```python
# Get the first derivative Ix
Ix = signal.convolve2d(my_img_gray, sobelx, "same")
Iy = signal.convolve2d(my_img_gray, sobely, "same")
```

```python
##----- Sobel
#-- Find the sobel operator Sobel=sqrt( Ix*Ix + Iy*Iy )

Sobel = np.sqrt((Ix*Ix) + (Iy*Iy))

plt.imshow(Sobel, cmap='gray')
```
