```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# generate the sobel filters  
sanka = plt.imread("Sanka_img.jpg")
sanka0 = np.rot90(sanka, k=-1)
sanka_gray = np.dot(sanka0[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(sanka_gray, cmap='gray')
```

```python
# x-kernel 
sobelx = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])

#y-kernel
sobely = np.array([[-1, -2, -1],[ 0,  0,  0],[ 1,  2,  1]])
```

```python
# Get the first derivative Ix
Ix = signal.convolve2d(sanka_gray, sobelx, "same")
Iy = signal.convolve2d(sanka_gray, sobely, "same")
```

```python
##----- Sobel
#-- Find the sobel operator Sobel=sqrt( Ix*Ix + Iy*Iy )

Sobel = np.sqrt( (Ix*Ix) + (Iy*Iy) )

plt.imshow(Sobel, cmap='gray')
```
