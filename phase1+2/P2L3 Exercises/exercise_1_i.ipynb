```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image  as matImage
#from PIL import Image
```

```python
# Using Matplotlib

# Read image
#image = matImage.imread('cheetah.jpg')
cat_img = plt.imread('images/cat.jpg')

# Display image
plt.imshow(cat_img)
```

```python
# show the type and shape of the image
type(cat_img)
```

```python
# Show the image shape
cat_img.shape
```

```python
# show the image array
cat_img
```

```python
# show the red color components
cat_img1 = cat_img.copy()
cat_img1[:, :, (0, 1)] = 0
plt.imshow(cat_img1)
```

```python
# Convert to grayscale
# rough method
img_grayscale = np.mean(cat_img, axis=-1)
# more accurate
gray = np.dot(cat_img[...,:3], [0.2989, 0.5870, 0.1140])

# Display grayscale image
#plt.imshow(img_grayscale, cmap='gray')
plt.imshow(gray, cmap='gray')
```

```python
# check the new shape
img_grayscale.shape
```

```python
# show the pixel value at the centre (256,320)
img_grayscale[256,320]
```

```python
# insert a white square at the center
img_grayscale[230:282, 300:340] = 255
plt.imshow(img_grayscale, cmap='gray')
```

```python
plt.imsave('images/cat_save0.jpg', img_grayscale, cmap='gray')
```

```python
# Using PIL


from PIL import Image
import numpy as np

pImg = np.asarray(Image.open('images/cat.jpg'))

# Display image
plt.imshow(pImg)

print(type(pImg))
# <class 'numpy.ndarray'>

print(pImg.dtype)
# uint8

print(pImg.shape)
# (225, 400, 3)
```

```python
pImg_gray = np.array(Image.open('images/cat.jpg').convert('L'))

# Display image
plt.imshow(pImg_gray,  cmap='gray')
print(pImg_gray.shape)
# (225, 400)
```

```python
# insert a white square at the center
pImg_gray[230:282, 300:340] = 255
plt.imshow(pImg_gray, cmap='gray')
```

```python
pil_img = Image.fromarray(pImg_gray)

pil_img.save('images/cat_save.jpg')
```
